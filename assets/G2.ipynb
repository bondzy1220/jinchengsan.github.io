```python
import pandas as pd
import numpy as np

df = pd.read_csv('G3D.csv')
df.head()

import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
Y =(df['latitude']) # 每一个点的X值
X =(df['longitude']) # 每一个点的Y值

plt.scatter(X, Y, s=1, alpha=1)

plt.show(10)
```

```python
rd=pd.read_csv('17S1.csv')
rd.head()


```

```python
import numpy as np
from matplotlib import pyplot as plt

Type_register = list(rd['Member Type']).count('Registered')
Type_casual = list(rd['Member Type']).count('Casual')
TT = Type_register + Type_casual

plt.figure(figsize=(6,9))
labels = ['member','casual']

sizes = [(Type_register/TT),(Type_casual/TT)]
colors = ['red','lightskyblue']

explode = (0.05,0)

patches,l_text,p_text = plt.pie(sizes,labels=labels,explode=explode, colors=colors,
                                labeldistance = 1.1,autopct = '%3.1f%%',shadow = False,
                                startangle = 90,pctdistance = 0.6)


for t in l_text:
    t.set_size=(30)
for t in p_text:
    t.set_size=(20)
plt.axis('equal')
plt.legend()
plt.show(0)
```
