```python

import pandas as pd
import numpy as np

df = pd.read_csv('17S1.csv')
df.head()
```

```python
ss= set(df['Start station'])


ee= set(df['End station'])

#ss==ee

SSS=list(ss)
#print(SSS[1])

from geopy.geocoders import Nominatim
geolocator= Nominatim()
#location=geolocator.geocode(SSS[0])

latitude=[]
longitude=[]
address=[]


S1 = SSS[:150]
S2 = SSS[150:300]
S3 = SSS[300:452]
#print(S1)
#print(S3)


for i in S1:
    location=geolocator.geocode(i)
    if location == None:
        address.append('NNNNNNNNN')
        latitude.append('NNNNNNNNN')
        longitude.append('NNNNNNNNN')
        #print('NNNNN')
    else:
        
        address.append(location.address)
        latitude.append(location.latitude)
        longitude.append(location.longitude)     
        #print(location.latitude)

for y in S2:
    location=geolocator.geocode(y)
    if location == None:
        address.append('NNNNNNNNN')
        latitude.append('NNNNNNNNN')
        longitude.append('NNNNNNNNN')
        #print('NNNNN')
    else:
        
        address.append(location.address)
        latitude.append(location.latitude)
        longitude.append(location.longitude)     
        #print(location.latitude)

for k in S3:
    location=geolocator.geocode(k)
    if location == None:
        address.append('NNNNNNNNN')
        latitude.append('NNNNNNNNN')
        longitude.append('NNNNNNNNN')
        #print('NNNNN')
    else:
        
        address.append(location.address)
        latitude.append(location.latitude)
        longitude.append(location.longitude)     
        #print(location.latitude)

print(latitude)

#print(location.address)
#print(location.latitude,location.longitude)
#print(location.raw)
```

```python
station_list=list(ss)

dataframe = pd.DataFrame({'K1':station_list,'K2':address,'K3':latitude,'K4':longitude})



dataframe.to_csv("S_L.csv")
```
